# Uniform prior

```python
def logprior(pars):
    if pars[0]<0 or pars[0]>10 or pars[1]<0 or pars[1]>10:
        return -np.inf
    return 0.0
```

```python
def loglikelihood(pars,data):
    return np.sum(beta.logpdf(data,pars[0],pars[1]))
 
    
```

```python
test=[a,b]
loglikelihood(test,data)
```

```python
def logpostpdf(pars,data):
    lp=logprior(pars)
    if not np.isfinite(lp):
        return -np.inf
    ll=loglikelihood(pars,data)# to put after the check

    return lp+ll
    
```

# MCMC


```python
ndim, nwalkers, nstep, nburn = 2, 20, 1000, 100 
firstguess=np.array([5,5])
p0 = [firstguess + 0.001*np.random.randn(ndim)  for i in range(nwalkers)]#first value of the parameters need to be close for each chain
p0
```

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, logpostpdf, args=[data]);#"ensemble" is good for low dimension
%time sampler.run_mcmc(p0, nstep);
```
