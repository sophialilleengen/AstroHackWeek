```python
from scipy.stats import beta, binom
import matplotlib.pyplot as plt
import numpy as np
import emcee
import corner
%matplotlib inline
```

```python
a = 1.5
b = 1.5

x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)

plt.plot(x, beta.pdf(x, a, b))
```

```python
#### Generate random numbers from beta distribution ####

r = beta.rvs(a, b, size=1000)
```

```python
plt.hist(r, normed=1, bins = 20)
plt.plot(r, beta.pdf(r, a, b), '.')
plt.show()
```

Bayes: $P(a, b| s) = P(s|a,b)P(a,b)/P(s)$

$ P (s|a,b) = $ beta distribution

```python
def log_prior(theta):
    if theta[0]<0 or theta[0]>10 or theta[1]<0 or theta[1]>10:
        return -np.inf
    return np.log(np.random.random())
        
def log_likelihood(theta, x):
    a, b = theta
    log_likelihood = np.sum(beta.logpdf(x, a, b)) 
    return log_likelihood

def log_posterior(theta, x):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x)
```

```python
ndim = 2  # number of parameters in the model
nwalkers = 20  # number of MCMC walkers
nburn = 100  # "burn-in" period to let chains stabilize
nsteps = 3000  # number of MCMC steps to take

firstguess=np.array([5, 5])
p0 = [firstguess + 0.001*np.random.randn(ndim)  for i in range(nwalkers)]#first value of the parameters need to be close for each chain

#starting_guesses = np.random.random((nwalkers, ndim))

#p0
```

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_posterior, args=[r]);
%time sampler.run_mcmc(p0, nsteps)
print("done")
```

```python
emcee_trace = sampler.chain[:, nburn:, :].reshape(-1, ndim).T
```
